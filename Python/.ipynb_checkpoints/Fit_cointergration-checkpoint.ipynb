```python
import pandas as pd
import numpy as np
import math
from statsmodels.tsa.stattools import coint
import statsmodels.api as sm
from matplotlib import pyplot as plt
import datetime
```

```python
data_EU = pd.read_csv("Data/EXUSEU.csv")
tsValues_EU = data_EU['EXUSEU']
tsValuesLog_EU = [math.log(1/ele) for ele in tsValues_EU]
data_CA = pd.read_csv("Data/EXCAUS.csv")
data_JP = pd.read_csv("Data/EXJPUS.csv")
data_CH = pd.read_csv("Data/EXCHUS_whole.csv")
data_KO = pd.read_csv("Data/EXKOUS.csv")
tsValuesLog_CA = [math.log(ele) for ele in data_CA['EXCAUS']]
tsValuesLog_JP = [math.log(ele) for ele in data_JP['EXJPUS']]
tsValuesLog_KO = [math.log(ele) for ele in data_KO['EXKOUS']]
tsValuesLog_CH = [math.log(ele) for ele in data_CH['EXCHUS']]
```

```python
from statsmodels.tsa.stattools import adfuller
```

```python
adfuller(tsValuesLog_CA)
```

```python
tsValuesLog_d_1_CA = [0]
tsValuesLog_d_1_CA += [tsValuesLog_CA[i+1]-tsValuesLog_CA[i] for i in range(172)]
```

```python
adfuller(tsValuesLog_d_1_CA,maxlag=1)
```

```python
output = pd.DataFrame(columns=['EU','CA','JP','CH','KO'],index=['EU','CA','JP','CH','KO'])
```

```python
adfuller(tsValuesLog_CH)
```

```python
tsValuesLog_d_1_CH = [0]
tsValuesLog_d_1_CH += [tsValuesLog_CH[i+1] - tsValuesLog_CH[i] for i in range(172)]
```

```python
adfuller(tsValuesLog_d_1_CH,maxlag=1)
```

```python
adfuller(tsValuesLog_EU)
```

```python
adfuller(tsValuesLog_JP,maxlag=1)
```

```python
tsValuesLog_d_1_JP = [0]
tsValuesLog_d_1_JP += [tsValuesLog_JP[i+1] - tsValuesLog_JP[i] for i in range(172)]
```

```python
adfuller(tsValuesLog_d_1_JP)
```

```python
adfuller(tsValuesLog_KO,maxlag=1)
```

```python
tsValuesLog_d_1_KO = [0]
tsValuesLog_d_1_KO += [tsValuesLog_KO[i+1] - tsValuesLog_KO[i] for i in range(172)]
```

```python
adfuller(tsValuesLog_d_1_KO,maxlag=1)
```

```python
ts = [tsValuesLog_EU,tsValuesLog_CA,tsValuesLog_JP,tsValuesLog_CH,tsValuesLog_KO]
```

```python
for i in range(5):
    for j in range(5):
        if i==j:
            output.iloc[i,j] = 0
            continue
        result = coint(ts[i],ts[j])
        output.iloc[i,j] = result[1]
```
